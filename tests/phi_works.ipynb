```python
from brian2 import *
import cPickle
```

```python
%pylab inline
```

```python
def data(al, beta, r_max, U):
    return cPickle.load(open('phis_al_{0}_b_{1}_rm_{2}_U_{3}.p'.format(al, beta, r_max, U)))
```

```python
alphas = [-50.0, -55.0]
betas = [0.2, 0.5]
r_maxs = [0.1, 0.3]
Us = arange(-70.0, -31.0, 5.0)
```

```python
dt=data(alphas[0], betas[0], r_maxs[0], Us[0])
```

```python
dt[1][0].shape[0]
```

```python
def phi(alpha, beta, r_max, us):
    return r_max/(1.0+exp(-beta*(us-alpha)))
```

```python
us = linspace(-70.0, -30.0, 1000)

for al in alphas:
    for b in betas:
        for r in r_maxs:
            phis = phi(al, b, r, us)
            figure()
            plot(us, phis)
            for U in Us:
                dt=data(al, b, r, U)
                assert(allclose(U, dt[0][0]/mV))
                spks = 1.0*dt[1][0].shape[0]
                scatter([U], [spks/60/1e3], s=50, lw=0, c='black')
```
